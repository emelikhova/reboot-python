```python
import requests
from tqdm.auto import tqdm
```

```python
sber = '3529'
page = 1
num_per_page = 100
moscow = 1
url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}&per_page={num_per_page}&area={moscow}' 
# f - ф-строка(т.е строка ,которая будет форматироваться  )
#  result - https://api.hh.ru/vacancies?employer_id=3529&page=1&per_page=100&area=1
```

```python
response = requests.get(url)
vacancies = response.json()
```

```python
print(vacancies.keys())
print(vacancies.get('page')) # ~ vacancies['page']
```

```python
num_pages = vacancies.get('pages')
num_pages
```

```python
all_vacancy_ids = []
for page in tqdm(range(num_pages)):
    url = f'https://api.hh.ru/vacancies?employer_id={sber}&page={page}&per_page={num_per_page}&area={moscow}' 
    response = requests.get(url)
    vacancies = response.json()
    vacancy_ids = [vacancy.get('id') for vacancy in vacancies.get('items')]
    all_vacancy_ids.extend(vacancy_ids)
```
